```python
#Run this cell
class Node:
  def __init__(self,val,next = None):
    self.val,self.next = val,next

def createList(arr):
  head = Node(arr[0])
  tail = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    tail.next = newNode
    tail = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

```python
def spiralMatrix(row,col,head):
  # Write your code here.
  temp=head
  r=row
  c=col
  count=0
  z=0
  matrix=np.array((row,col),dtype=int)
  idx_em=(row*col)
  while temp!= None:
    matrix[r][c]=temp.elem
    temp=temp.next
    count+=1

  for i in range(col):
    while temp!=None:
      p=0
      matrix[p][i]=temp.elem
      temp=temp.next
      z+=1
      idx_em=(row+col)-1


  for i in range(row):
    while temp!=None:
      if z!=count and idx_em!=0:
        matrix[i][c]=temp.elem
        temp=temp.next
        z+=1
        idx_em=(row*col)-1

  for i in range(col-1,-1,-1):
    while temp!=None:
      if z!=count and idx_em!=0:
        matrix[r][i]=temp.elem
        temp=temp.next
        z+=1
        idx_em=(row*col)-1
  for i in range(row-1,0,1):
      while temp!=None:
        if z!=count and idx_em!=0:
        matrix[r][i]=temp.elem
        temp=temp.next
        z+=1
        idx_em=(row*col)-1
  return matrix






head1 = createList(np.array([3,0,2,6,8,1,7,9,4,2,5,5,0]))
matrix1 = spiralMatrix(3,5,head1)
print('Output --- ')
print(matrix1)
# The output should be [[3,0,2,6,8],[5,0,-1,-1,1],[5,2,4,9,7]]
print('____________________________________')
head2 = createList(np.array([0,1,2]))
matrix2 = spiralMatrix(1,5,head2)
print('Output --- ')
print(matrix2)
# The output should be  [[0,1,2,-1]]
```
